```python
import cv2
img = cv2.imread('cherprang48.jpg')
```

```python
rgb_img = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.figure(figsize=(5, 5))
plt.imshow(rgb_img)
```

```python
roi = rgb_img[100:800, 600:1400]
plt.figure(figsize=(3, 3))
plt.imshow(roi)
```

```python
new_img = rgb_img.copy()
cv2.rectangle(new_img, (600, 200), (1300, 800), (255, 0, 0), 5)
plt.figure(figsize=(5, 5))
plt.imshow(rgb_img)
```

```python
cv2.putText(new_img, 'Sherprang', (300, 400), cv2.FONT_HERSHEY_COMPLEX, 5, (0, 255, 0), 5)
plt.figure(figsize=(5, 5))
plt.imshow(new_img)
```
